```python
## Identify basic geometries using contours approximation
```

```python
import cv2
import numpy as np

#Load image and than convert it to grayscale

image = cv2.imread('images/someshapes.jpg')
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

cv2.imshow('Identifying Geometries',image)

cv2.waitKey()

ret,thresh = cv2.threshold(gray,127,255,1)

#Extracting contours
contours,hierarcy = cv2.findContours(thresh.copy(),cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)

for i in contours:
    
    approx  = cv2.approxPolyDP(i,0.01*cv2.arcLength(i,True),True)
    
    if len(approx) == 3:
        shape_name = 'Triangle'
        cv2.drawContours(image,[i],0,(0,255,0),-1)
        
        #Find contour center to place text right at the center
        M = cv2.moments(i)
        cx = int(M['m10']/M['m00'])
        cy = int(M['m01']/M['m00'])
        cv2.putText(image,shape_name,(cx-50,cy),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1)
    
    elif len(approx) == 4:
        x,y,w,h = cv2.boundingRect(i)

        #Check to see if 4 side polygon is square or rectangle
        #cv2.bounding rects returns the top left then width and
        if abs(w-h) <= 3:
            shape_name = 'Square'
            
            #Find the contour center to place text
            M = cv2.moments(i)
            cx = int(M['m10']/M['m00'])
            cy = int(M['m01']/M['m00'])
        
            cv2.drawContours(image,[i],0,(0,125,255),-1)
            cv2.putText(image,shape_name,(cx-50,cy),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1)
        else:
            shape_name = 'Rectangle'
            
            #Find the contour center to place text
            cv2.drawContours(image,[i],0,(0,0,255),-1)
            M = cv2.moments(i)
            cx = int(M['m10']/M['m00'])
            cy = int(M['m01']/M['m00'])
            cv2.putText(image,shape_name,(cx-50,cy),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1)  
    elif len(approx) == 10:
        
        shape_name ='Star'
        cv2.drawContours(image,[i],0,(255,255,0),-1)
        #Find contour center to place text right at the center
        M = cv2.moments(i)
        cx = int(M['m10']/M['m00'])
        cy = int(M['m01']/M['m00'])
        cv2.putText(image,shape_name,(cx-50,cy),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1)
    
    elif len(approx) >=15:
        
        shape_name = 'Circle'
        cv2.drawContours(image,[i],0,(0,255,255),-1)
        M = cv2.moments(i)
        cx = int(M['m10']/M['m00'])
        cy = int(M['m01']/M['m00'])
        cv2.putText(image,shape_name,(cx-50,cy),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1)
        
    cv2.imshow('Identfying shapes',image)
    
    cv2.waitKey(0)
    
cv2.destroyAllWindows()
    
            
        
```
